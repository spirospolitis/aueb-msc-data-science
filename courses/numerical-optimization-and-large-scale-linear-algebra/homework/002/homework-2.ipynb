## Problem 1

Show that equation

$$
min_{\substack{\mathbf{f}}}
\left \{
\lVert 
\mathbf{g} - \mathbf{K} \mathbf{f}
\rVert_{2}^{2} +
\alpha^{2}
\lVert 
\mathbf{f}
\rVert_{2}^{2}
\right \}
\;\;\;\;\;\;\;\;\;\; \tag{1.1}
$$

is equivalent to the linear least-squares problem

$$
min_{\substack{\mathbf{f}}}
\left \lVert 
\begin{bmatrix}
    \mathbf{g} \\
    0
\end{bmatrix} -
\begin{bmatrix}
    \mathbf{K} \\
    \alpha \mathbf{I}
\end{bmatrix} \mathbf{f}
\right \rVert_{2}^{2}
\;\;\;\;\;\;\;\;\;\; \tag{1.2}
$$

### Answer

Viewing $(1.2)$ represented as

$$
\begin{bmatrix}
    \mathbf{g} \\
    0
\end{bmatrix} =
\begin{bmatrix}
    \mathbf{x_{1}} \\
    0
\end{bmatrix}
$$

and 

$$
\begin{bmatrix}
    \mathbf{K} \mathbf{f} \\
    \alpha \mathbf{f}
\end{bmatrix} =
\begin{bmatrix}
    \mathbf{y_{1}} \\
    \alpha \mathbf{y_{2}}
\end{bmatrix}
$$

with $x_{1}, y_{1} \in R^{l}$, $y_{2} \in R^{m}$, such that $l + m = n$, and $\alpha$ a scalar, we get the following:

$$
\left \lVert 
\begin{bmatrix}
    \mathbf{x_{1}} \\
    0
\end{bmatrix} -
\begin{bmatrix}
    \mathbf{y_{1}} \\
    \alpha \mathbf{y_{2}}
\end{bmatrix}
\right \rVert_{2}^{2} =
$$

$$
\left \lVert 
\begin{bmatrix}
    \mathbf{x_{1}} - \mathbf{y_{1}} \\
    - \alpha \mathbf{y_{2}}
\end{bmatrix}
\right \rVert_{2}^{2} =
$$

$$
\left(
\sqrt{(\mathbf{x_{1}} - \mathbf{y_{1}})^{2} + (-\alpha \mathbf{y_{2}})^{2}}
\right)^{2} =
$$

$$
(\mathbf{x_{1}} - \mathbf{y_{1}})^{2} + (-\alpha \mathbf{y_{2}})^{2} =
$$

$$
(\mathbf{x_{1}} - \mathbf{y_{1}})^{2} + \alpha^{2} \mathbf{y_{2}}^{2} =
$$

(substituting $\mathbf{x_{1}}$, $\mathbf{y_{1}}$, $\mathbf{y_{2}}$ and because they are vectors)

$$
\left \lVert 
\mathbf{g} - \mathbf{Kf}
\right \rVert_{2}^{2} +
\alpha^{2}
\left \lVert 
\mathbf{f}
\right \rVert_{2}^{2}
\;\;\;\;\;\;\;\;\;\; \tag{1.3}
$$

Clearly, $(1.3)$ is equal to $(1.1)$.

## Problem 2

Show that if $K$ has a singular value decomposition $\mathbf{K} = \mathbf{U} \mathbf{\Sigma} \mathbf{V^{T}}$, then equation 

$$
min_{\substack{\mathbf{f}}}
\left \lVert 
\begin{bmatrix}
    \mathbf{g} \\
    0
\end{bmatrix} -
\begin{bmatrix}
    \mathbf{K} \\
    \alpha \mathbf{I}
\end{bmatrix} \mathbf{f}
\right \rVert_{2}^{2}
\;\;\;\;\;\;\;\;\;\; \tag{2.1}
$$

can be transformed to the equivalent least squares problem

$$
min_{\substack{\mathbf{g}}}
\left \lVert 
\begin{bmatrix}
    \mathbf{\hat{g}} \\
    0
\end{bmatrix} -
\begin{bmatrix}
    \mathbf{\Sigma} \\
    \alpha \mathbf{I}
\end{bmatrix} \mathbf{\hat{f}}
\right \rVert_{2}^{2}
\;\;\;\;\;\;\;\;\;\; \tag{2.2}
$$

where $\mathbf{\hat{f}} = \mathbf{V^{T}} \mathbf{f}$ and $\mathbf{\hat{g}} = \mathbf{U^{T}} \mathbf{g}$.

### Answer

From problem 1, eq. $(1.3)$, we can deduce that $(2.2)$ can be written as:

$$
\left \lVert 
\mathbf{\hat{g}} - \mathbf{\Sigma} \mathbf{\hat{f}}
\right \rVert_{2}^{2} +
\alpha^{2}
\left \lVert 
\mathbf{\hat{f}}
\right \rVert_{2}^{2} =
$$

$$
(\mathbf{\hat{g}} - \mathbf{\Sigma} \mathbf{\hat{f}})^{T} (\mathbf{\hat{g}} - \mathbf{\Sigma} \mathbf{\hat{f}}) + \alpha^{2} \mathbf{\hat{f}}^{T} \mathbf{\hat{f}} =
$$

$$
\mathbf{\hat{g}}^{T} \mathbf{\hat{g}} - \mathbf{\hat{g}}^{T} \mathbf{\Sigma} \mathbf{\hat{f}} - (\mathbf{\Sigma} \mathbf{\hat{f}})^{T} \mathbf{\hat{g}} + (\mathbf{\Sigma} \mathbf{\hat{f}})^{T} \mathbf{\Sigma} \mathbf{\hat{f}} + \alpha^{2} \mathbf{\hat{f}}^{T} \mathbf{\hat{f}} =
$$

$$
\mathbf{\hat{g}}^{T} \mathbf{\hat{g}} - 2\mathbf{\hat{g}}^{T} \mathbf{\Sigma} \mathbf{\hat{f}} + \mathbf{\hat{f}}^{T} \mathbf{\Sigma}^{T} \mathbf{\Sigma} \mathbf{\hat{f}} + \alpha^{2} \mathbf{\hat{f}}^{T} \mathbf{\hat{f}}
\;\;\;\;\;\;\;\;\;\; \tag{2.3}
$$

Breaking up $(2.3)$ we get the following equations and setting $\mathbf{\hat{f}} = \mathbf{V^{T}} \mathbf{f}$ and $\mathbf{\hat{g}} = \mathbf{U^{T}} \mathbf{g}$ we get the following:

$$
\mathbf{\hat{g}}^{T} \mathbf{\hat{g}} = 
(\mathbf{U}^{T} \mathbf{g})^{T} \mathbf{U}^{T} \mathbf{g} = 
\mathbf{g}^{T} \mathbf{U} \mathbf{U}^{T} \mathbf{g} =
\mathbf{g}^{T} \mathbf{g}
\;\;\;\;\;\;\;\;\;\; \tag{2.4}
$$

Note: on $(2.4)$ $\mathbf{U} \mathbf{U}^{T} = \mathbf{I}$, because $\mathbf{U}$ is orthogonal.

$$
-2 \mathbf{\hat{g}}^{T} \mathbf{\Sigma} \mathbf{\hat{f}} = 
-2 (\mathbf{U}^{T} \mathbf{g})^{T} \mathbf{\Sigma} \mathbf{V}^{T} \mathbf{f} = 
-2 \mathbf{g}^{T} \mathbf{U} \mathbf{\Sigma} \mathbf{V}^{T} \mathbf{f} = 
-2 \mathbf{g}^{T} \mathbf{K} \mathbf{f}
\;\;\;\;\;\;\;\;\;\; \tag{2.5}
$$

$$
\mathbf{\hat{f}}^{T} \mathbf{\Sigma}^{T} \mathbf{\Sigma} \mathbf{\hat{f}} = 
(\mathbf{V}^{T} \mathbf{f})^{T} \mathbf{\Sigma}^{T} \mathbf{\Sigma} \mathbf{V}^{T} \mathbf{f} =
\mathbf{f}^{T} \mathbf{V} \mathbf{\Sigma}^{T} \mathbf{\Sigma} \mathbf{V}^{T} \mathbf{f} = 
\mathbf{f}^{T} \mathbf{V} \mathbf{\Sigma}^{T} \mathbf{U}^{T} \mathbf{U} \mathbf{\Sigma} \mathbf{V}^{T} \mathbf{f}  = 
\mathbf{f}^{T} \mathbf{Κ}^{T} \mathbf{Κ} \mathbf{f}
\;\;\;\;\;\;\;\;\;\; \tag{2.6}
$$

Note: on $(2.6)$ we have set $\mathbf{U}^{T} \mathbf{U}$ on step 4, as a substitution for $\mathbf{I}$, since $\mathbf{U}^{T} \mathbf{U} = \mathbf{I}$. We did this because we know that $\mathbf{K} = \mathbf{U} \mathbf{\Sigma} \mathbf{V}^{T}$ and we need to arrive to such a factorization that will allow us to substitute $\mathbf{U} \mathbf{\Sigma} \mathbf{V}^{T}$ with $\mathbf{K}$.

$$
\alpha^{2} \mathbf{\hat{f}}^{T} \mathbf{\hat{f}} = 
\alpha^{2} (\mathbf{V}^{T} \mathbf{f})^{T} \mathbf{V}^{T} \mathbf{f} =
\alpha^{2} \mathbf{f}^{T} \mathbf{V} \mathbf{V}^{T} \mathbf{f} =
\alpha^{2} \mathbf{f}^{T} \mathbf{f}
\;\;\;\;\;\;\;\;\;\; \tag{2.7}
$$

Note: on $(2.7)$ $\mathbf{V} \mathbf{V}^{T} = \mathbf{I}$, because $\mathbf{V}$ is orthogonal.

Substituting $(2.4)$, $(2.5)$, $(2.6)$ and $(2.7)$ back into $(2.3)$, we get:

$$
\mathbf{g}^{T} \mathbf{g} -2 \mathbf{g}^{T} \mathbf{K} \mathbf{f} + \mathbf{f}^{T} \mathbf{Κ}^{T} \mathbf{Κ} \mathbf{f} + \alpha^{2} \mathbf{f}^{T} \mathbf{f} =
(\mathbf{g} -\mathbf{K} \mathbf{f})^{T} (\mathbf{g} -\mathbf{K} \mathbf{f}) + \alpha^{2} \mathbf{f}^{T} \mathbf{f} =
\left \lVert 
\mathbf{g} - \mathbf{Kf}
\right \rVert_{2}^{2} +
\alpha^{2}
\left \lVert 
\mathbf{f}
\right \rVert_{2}^{2}
\;\;\;\;\;\;\;\;\;\; \tag{2.8}
$$

Since we know from problem 1 that $(2.8)$ is equivalent to $(1.2)$, we have shown that $(2.2)$ is equivalent to $(2.1)$.

## Problem 3

Determine a formula for the solution of equation

$$
min_{\substack{\mathbf{g}}}
\left \lVert 
\begin{bmatrix}
    \mathbf{\hat{g}} \\
    0
\end{bmatrix} -
\begin{bmatrix}
    \mathbf{\Sigma} \\
    \alpha \mathbf{I}
\end{bmatrix} \mathbf{\hat{f}}
\right \rVert_{2}^{2}
\;\;\;\;\;\;\;\;\;\; \tag{2.2}
$$

where $\mathbf{\hat{f}} = \mathbf{V^{T}} \mathbf{f}$ and $\mathbf{\hat{g}} = \mathbf{U^{T}} \mathbf{g}$.

Hint: you should set the drivative of the minimization function to zero and solve for $\mathbf{\hat{f}}$.

### Answer

From $(2.3)$ we have:

$$
min_{\substack{\mathbf{g}}}
\left \lVert 
\mathbf{\hat{g}} - \mathbf{\Sigma} \mathbf{\hat{f}}
\right \rVert_{2}^{2} +
\alpha^{2}
\left \lVert 
\mathbf{\hat{f}}
\right \rVert_{2}^{2} =
$$

$$
min_{\substack{\mathbf{g}}}
\left(
\left(
\sqrt{\sum_{i=1}^{n} (\hat{g}_{i} - \sigma_{i} \hat{f}_{i} )^{2}}
\right)^{2} +
\alpha^{2}
\left(
\sqrt{\sum_{i=1}^{n} \hat{f}_{i}^{2}}
\right)^{2}
\right) =
$$

$$
min_{\substack{\mathbf{g}}}
\left(
\sum_{i=1}^{n} (\hat{g}_{i} - \sigma_{i} \hat{f}_{i} )^{2} +
\alpha^{2}
\sum_{i=1}^{n} \hat{f}_{i}^{2}
\right)
\;\;\;\;\;\;\;\;\;\; \tag{3.1}
$$

Taking the partial derivative of $(3.1)$ w.r.t. $\hat{f}$ we have:

$$
\frac{\partial}{\partial \hat{f}} = 2(\hat{g}_{i} - \sigma_{i} \hat{f}_{i})(-\sigma_{i}) + 2 \alpha^{2} \hat{f}_{i} = 
-2 \sigma_{i} \hat{g}_{i} + 2 \sigma_{i}^{2} \hat{f}_{i} + 2 \alpha^{2} \hat{f}_{i}
\;\;\;\;\;\;\;\;\;\; \tag{3.2}
$$

Setting $(3.2)$ to zero and solving for $\hat{f}_{i}$ we get:

$$
-2 \sigma_{i} \hat{g}_{i} + 2 \sigma_{i}^{2} \hat{f}_{i} + 2 \alpha^{2} \hat{f}_{i} = 0 \implies
$$

$$
- \sigma_{i} \hat{g}_{i} + \sigma_{i}^{2} \hat{f}_{i} + \alpha^{2} \hat{f}_{i} = 0 \implies
$$

$$
\hat{f}_{i} (\sigma_{i}^{2} + \alpha^{2}) = \sigma_{i} \hat{g}_{i} \implies
$$

$$
\hat{f}_{i} = \frac{\sigma_{i} \hat{g}_{i}}{\sigma_{i}^{2} + \alpha^{2}}
\;\;\;\;\;\;\;\;\;\; \tag{3.3}
$$

## Problem 4

Show that the solution to the problem

$$
min_{\substack{\mathbf{f}}}
\left \{
\lVert 
\mathbf{g} - \mathbf{K} \mathbf{f}
\rVert_{2}^{2} 
\right \}
\;\;\;\;\;\;\;\;\;\; \tag{4.1}
$$

is

$$
\mathbf{f}_{ls} =
\mathbf{V} \mathbf{\Sigma}^{\dagger} \mathbf{U}^{T} \mathbf{g} = 
\displaystyle\sum_{i=1}^{n} \frac{\mathbf{u}_{i}^{T} \mathbf{g}}{\sigma_{i}} \mathbf{v}_{i}
\;\;\;\;\;\;\;\;\;\; \tag{4.2}
$$

where $u_{i}$ is the ith column column of $\mathbf{U}$ and $\mathbf{v}_{i}$ is the ith column of $\mathbf{V}$.

### Answer

Let's visualize the matrices, vectors and their operations, as stated in the problem:

$$
\mathbf{U} = 
\begin{bmatrix}
    \vert & \vert & \cdots & \vert \\
    u_{1} & u_{2} & \cdots & u_{n} \\
    \vert & \vert & \cdots & \vert
\end{bmatrix},
\mathbf{U}^{T} = 
\begin{bmatrix}
    \text{___} & u_{1}^{T} & \text{___} \\
    \text{___} & u_{2}^{T} & \text{___} \\
    \vdots & \vdots & \vdots \\
    \text{___} & u_{n}^{T} & \text{___} \\
\end{bmatrix}
$$

$$
\mathbf{V} = 
\begin{bmatrix}
    \vert & \vert & \cdots & \vert \\
    v_{1} & v_{2} & \cdots & v_{n} \\
    \vert & \vert & \cdots & \vert
\end{bmatrix},
\mathbf{V}^{T} = 
\begin{bmatrix}
    \text{___} & v_{1}^{T} & \text{___} \\
    \text{___} & v_{2}^{T} & \text{___} \\
    \vdots & \vdots & \vdots \\
    \text{___} & v_{n}^{T} & \text{___} \\
\end{bmatrix}
$$

From problem $(2)$, we know that:

$$
\mathbf{\hat{g}} = \mathbf{U^{T}} \mathbf{g} \implies \hat{g}_{i} = \mathbf{u}_{i}^{T} \mathbf{g}
\;\;\;\;\;\;\;\;\;\; \tag{4.3}
$$

$$
\mathbf{\hat{f}} = \mathbf{V^{T}} \mathbf{f} \implies 
\hat{f}_{i} = \mathbf{v}_{i}^{T} \mathbf{g}
\;\;\;\;\;\;\;\;\;\; \tag{4.4}
$$

From problem $(3)$, equation $(3.3)$, we expressed the least squares problem as:

$$
\hat{f}_{i} = \frac{\sigma_{i} \hat{g}_{i}}{\sigma_{i}^{2} + \alpha^{2}}
\;\;\;\;\;\;\;\;\;\; \tag{4.5}
$$

Since we are been asked to evaluate

$$
min_{\substack{\mathbf{f}}}
\left \{
\lVert 
\mathbf{g} - \mathbf{K} \mathbf{f}
\rVert_{2}^{2} 
\right \}
$$

we can infer that no regularization is considered, (i.e. for $\alpha = 0$). Hence the above becomes:

$$
\hat{f}_{i} = \frac{\sigma_{i} \hat{g}_{i}}{\sigma_{i}^{2}} \implies \hat{f}_{i} = \frac{\hat{g}_{i}}{\sigma_{i}} \implies
\hat{\mathbf{f}} = \sum_{i=1}^{n} \frac{\hat{g}_{i}}{\sigma_{i}}
\;\;\;\;\;\;\;\;\;\; \tag{4.6}
$$

Therefore, from $(4.3)$, $(4.4)$, $(4.6)$ and because $\mathbf{V}$ is orthogonal, we have:

$$
\mathbf{V}^{T} \mathbf{f} = \sum_{i=1}^{n} \frac{\mathbf{u}_{i}^{T} \mathbf{g}} {\sigma_{i}} \implies
$$

$$
\mathbf{f} = (\mathbf{V}^{T})^{-1} \sum_{i=1}^{n} \frac{\mathbf{u}_{i}^{T} \mathbf{g}} {\sigma_{i}} \implies
$$

$$
\mathbf{f} = \mathbf{V} \sum_{i=1}^{n} \frac{\mathbf{u}_{i}^{T} \mathbf{g}} {\sigma_{i}} \implies
$$

$$
\mathbf{f} = \sum_{i=1}^{n} \frac{\mathbf{u}_{i}^{T} \mathbf{g}} {\sigma_{i}} \mathbf{v}_{i}
$$

### Introductory problem statement and theory

The generic form of the problem we are called upon to solve is: 

$$\mathbf{g} = \mathbf{K} \mathbf{f} + \eta$$ 

where:

$\mathbf{g}$: the $n^{2}$ vector of observed image pixels.

$\mathbf{K}$: a $nxn$ ill-conditioned degradation matrix representing the blurring operator applied to the original, unknown image.

$\mathbf{f}$: approximation of the original image as a $n^{2}$ vector.

$\eta$: random noise

$\mathbf{A}$, $\mathbf{B}$ are given as inputs to the problem an they represent the operands of the Kronecker product $\mathbf{K} = \mathbf{A} \otimes \mathbf{B}$ that produces matrix $\mathbf{K}$. This product is too large to be used directly in SVD. In fact, it is of the order $n^{2} * n^{2} = n^{4}$. It quickly becomes evident that we need a more efficient method, by way of computing partial Kronecker products.

Utilising the following theorem:


If 

$$\mathbf{A} = \mathbf{U}_{A} \mathbf{\Sigma}_{A} \mathbf{V}_{A}^{T}$$

$$\mathbf{B} = \mathbf{U}_{B} \mathbf{\Sigma}_{B} \mathbf{V}_{B}^{T}$$

then

$$\mathbf{K} = \mathbf{U} \mathbf{\Sigma} \mathbf{V}^{T}$$

with
 
$$\mathbf{U} = \mathbf{U}_{A} \otimes \mathbf{U}_{B}$$
$$\mathbf{\Sigma} = \mathbf{\Sigma}_{A} \otimes \mathbf{\Sigma}_{B}$$
$$\mathbf{V} = \mathbf{V}_{A} \otimes \mathbf{V}_{B}$$

we can tell that, in order to compute the SVD of $\mathbf{K}$, we need to reconstruct $\mathbf{U}$, $\mathbf{\Sigma}$, $\mathbf{V}$ from the respective Kronecker products of the components of the SVDs of matrices $\mathbf{A}$, $\mathbf{B}$. In other words, the components of the SVDs of the given input matrices $\mathbf{A}$, $\mathbf{B}$ are the factors of the Kronecker products of $\mathbf{U}$, $\mathbf{\Sigma}$, $\mathbf{V}$.

Applying singular value decomposition to the matrix A, we have 

$$\mathbf{A} = \mathbf{U} \mathbf{\Sigma} \mathbf{V}^{T}$$

$\mathbf{\Sigma}$ is a diagonal matrix of singular values of the matrix $\mathbf{A}$, whose non-zero diagonal elements $\lambda_{1}, \lambda_{2}, \cdots , \lambda_{i}$ are, without loss of generality, assumed to be positive and arranged in decreasing order i.e. $\lambda_{1} \geq \lambda_{2} \geq \lambda_{i} \gt 0$.

### Truncated SVD

In order to compute the truncated SVD reconstruction of $\mathbf{F}$, we utilized the solution to problem $(4.1)$, specifically its equivalentl expression $(4.2)$.

To achieve the computation of $(4.2)$, we implemented the following process:

- Computed the quantities $\mathbf{U_{A}}$, $\mathbf{\Sigma_{A}}$, $\mathbf{V_{A}^{T}}$ by performing SVD on $\mathbf{A}$. Similarly, we computed the quantities $\mathbf{U_{B}}$, $\mathbf{\Sigma_{B}}$, $\mathbf{V_{B}^{T}}$ by performing SVD on $\mathbf{B}$.

- Computed the Kronecker product of ($\mathbf{\Sigma_{A}}$, $\mathbf{\Sigma_{B}}$), an efficient computation since both are vectors of size $n^{2}$. The reason for this computation is to retrieve the singular values of $\mathbf{K}$.

- The Kronecker product of ($\mathbf{\Sigma_{A}}$, $\mathbf{\Sigma_{B}}$) was then sorted in descending order, thereby allowing us to retrieve the $p$-first, most significant singular values of $\mathbf{K}$.

- Computing $\hat{f}$ is then achieved by applying $(4.2)$, making sure to select such columns of $\mathbf{U_{A}}$, $\mathbf{V_{A}}$ and $\mathbf{U_{B}}$, $\mathbf{V_{B}}$, such that they correspond to the ordered index of the Kronecker product of ($\mathbf{\Sigma_{A}}$, $\mathbf{\Sigma_{B}}$), as if $\mathbf{\Sigma}$ was represented in diagonal matrix form (not computationally feasible, since such a representation would yield a $n^{2} \times n^{2}$ matrix.).

In order to pick a good candidate value for $p$, we need to visualize the order singular values so that we can get a visual feel of their weight:

```python
# Compute SVD of A, B.
U_A, S_A, V_A_T = np.linalg.svd(A)
U_B, S_B, V_B_T = np.linalg.svd(B)

S = np.flip(np.sort(np.kron(S_A, S_B)))
```

### Tikhonov regularization (Regularized Least Squares optimization)

Recall that the problem we seek to solve, from $(1.1)$ is

$$
min_{\substack{\mathbf{f}}}
\left \{
\lVert 
\mathbf{g} - \mathbf{K} \mathbf{f}
\rVert_{2}^{2} +
\alpha^{2}
\lVert 
\mathbf{f}
\rVert_{2}^{2}
\right \}
$$

We seek a good value for parameters $\alpha$, such that it limits the size of the coefficients by adding a penalty equal to the L2 norm of the magnitude of coefficients (i.e. performing L2 regularization).

In effect, we seek to formulate a less noise-sensitive approach to the Least Squares problem.

In order to solve the problem, we will use its equivalent $(3.3)$ representation

$$
\hat{f}_{i} = \frac{\sigma_{i} \hat{g}_{i}}{\sigma_{i}^{2} + \alpha^{2}}
$$

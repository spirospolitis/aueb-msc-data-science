#### Answer 1.a.i.

##### Power method implementation, alpha = 0.85

As mentioned in [[1]](#ref-1), the *Power iteration* (also known as the *Power method*) is an eigenvalue algorithm: given a diagonalizable matrix $\mathbf{A}$, the algorithm will produce a number ${\lambda}$, which is the greatest (in absolute value) eigenvalue of $\mathbf{A}$, and a nonzero vector $\mathbf{v}$, the corresponding eigenvector of $\lambda$, such that $\mathbf{A} \mathbf{v} = \lambda \mathbf{v}$.

PageRank is determined by computing the stationary solution $\pi^{T}$ of the hyperlink structure of the web, represented as a Markov chain with a transition probability matrix $\mathbf{P}$. The row vector $\pi^{T}$ can be found by solving the eigenvector problem $\pi^{T} \overline{\overline{P}}= \pi^{T}$, which translates to user power iterations for expression (1) of par. 5.1 of [[2]](#ref-2).

Based on theory (see par. 3, pp. 3 and par. 5.1.1, pp.  of [[2]](#ref-2)), we expect the algorithm to converge in accordance with the rate

$$
\left| \frac{\lambda_{2}}{\lambda_{1}} \right|
\;\;\;\;\;\;\;\;\;\; \tag{1.a.i.1}
$$

where $\lambda_{1}$, $\lambda_{2}$ are the $1st$ and $2nd$ dominant eigenvalues, respectively.

```python
eigenvalues_1ai = np.sort(-pr_p_1ai.x.toarray())
lambda_1_1ai = eigenvalues_1ai[0, 0]
lambda_2_1ai = eigenvalues_1ai[0, 1]

conv_rate_1ai = np.round(lambda_2_1ai / lambda_1_1ai, 3)

md('Convergence rate of the Power method: $%s$.' % conv_rate_1ai)
```

As reported in pagerank.pdf, a rough estimate of the number of iterations needed to converge to a tolerance level $\tau$ is 

$$\frac{log_{10}(\alpha)}{log_{10}(\tau)}$$

In this case, for $\tau = 1e-08$ and $\alpha = 0.85$ we can expect approximately:

(source: Wikipedia article on *Power iteration*, https://en.wikipedia.org/wiki/Power_iteration)

```python
est_iter_1ai = np.log10(1e-08) / np.log10(0.85)

md('Expected iterations for the convergence of the Power method: $%s$.' % np.floor(est_iter_1ai))
```

#### Answer 1.a.ii

##### Gauss-Seidel implementation, alpha = 0.85

In numerical linear algebra, the *Gauss–Seidel* method (see [[3]](#ref-3)), also known as the *Liebmann* method or the *method of successive displacement*, is an iterative method used to solve a linear system of equations. Though it can be applied to any matrix with non-zero elements on the diagonals, convergence is only guaranteed if the matrix is either diagonally dominant, or symmetric and positive definite.

Our goal in this context (see [[5]](#ref-5)) is to compute the Gauss-Seidel iterations of the form:

$$
x_{i}^{(k+1)} = (1-\alpha) + \alpha \sum_{j<i} a_{ij} x_{j}^{(k+1)} + \alpha \sum_{j>i} a_{ij} x_{j}^{(k)} \text{   } \forall i
$$

Having tried multiples approache for the solution of the system, including iteratives ones, and having had very slow performance on iterative methods, I focused on an implementation that exploits the fact that the approximation to the solution of a generic linear system of equations $\mathbf{A} \mathbf{x} = \mathbf{b}$ can be written as:

$$
\mathbf{L} \mathbf{x}^{(k+1)} = \mathbf{b} - \mathbf{U} \mathbf{x}^{(k)}
$$

where $\mathbf{x}^{(k)}$ is the k-th approximation of $\mathbf{x}$, $\mathbf{x}^{(k+1)}$ is the next iteration of $\mathbf{x}$ and $\mathbf{P}$ can be decomposed to a lower triangular matrix $\mathbf{L}$ and a strictly upper triangular matrix $\mathbf{U}$, such that $\mathbf{P} = \mathbf{L} + \mathbf{U}$ (see par. "Description" of [[3]](#ref-3), par. 5.4, pp. 19 of [[4]](#ref-4)).


Therefore, I broke down the matrix $\mathbf{P}$ to a lower triangular matrix $\mathbf{L}$ and a stricty triangular matrix $\mathbf{U}$, satisfying the properties described above, thereafter solving the linear system with the solver provided by *Scipy* within a cycle of iterations, until convergence was achieved.

The computation cost of the implementation breaks down to:

- LU factorization: $O(1)$, since this factorization is simply braking down $\mathbf{P}$ to a sum of matrices $\mathbf{L}$ and $\mathbf{U}$.
- Solution of a triangular system:  $O(n^{2}) \times \text{count of linear systems}$

with a total cost of $O(1) + (O(n^{2}) \times \text{count of linear systems})$

In this context, an epoch is a set of iterations the algorithm performs until ultimately converging.

## Part (2)

A typical way to raise the PageRank of a page is to use ?link farms?, i.e., a collection of ?fake? pages that point to yours in order to improve its PageRank. Our goal in this problem is to do a little analysis of the design of link farms, and how their structure affects the PageRank calculations. Consider the web graph. It contains $n$ pages, labeled 1 through $n$: Of course, $n$ is very large. As mentioned in, we use the notation $G = \alpha P + \frac{1 − \alpha}{n} I$ for the transition matrix. Let $\pi_{i}$ i denote the PageRank of page $i$ and $\pi = (\pi_{1}, \pi_{2}, \cdots \pi_{n})$ denote the vector of PageRanks of all pages. Note: For a page that has $k$ outgoing links, we put 1 = $k$ for the corresponding entries of $P$: However, when a webpage has no outgoing links, we add a 1 as the corresponding diagonal element of $P$ for making its row-sum one. Note that this makes $G$ a valid transition probability matrix.

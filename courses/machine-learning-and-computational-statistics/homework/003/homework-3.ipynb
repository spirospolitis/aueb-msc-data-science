### Answer (a)

According to the definition of unbiased estimators, it should hold that $E[\theta'_{MVU}] = \theta_{0}$

### Answer (b)

In general, for an estimator to be biased, it must hold that:

$E[\hat{\theta}] - \theta \neq 0$, where $\hat{\theta}$ is the estimator and $\theta$ is the true population parameter.

Therefore we have:

$E[\theta'_{b}] - \theta_{0} = $
$E[(1 + \alpha) \cdot \theta'_{MVU}] - \theta_{0} =$
$E[\theta'_{MVU} + \alpha \cdot \theta'_{MVU}] - \theta_{0} = $
$E[\theta'_{MVU}] + E[\alpha \cdot \theta'_{MVU}] - \theta_{0} \textbf{ (1)}$

Because $E[\theta'_{MVU}]$ is an unbiased estimator of $\theta$ it holds that:

$E[\theta'_{MVU}] - \theta_{0} = 0 \textbf{ (2)}$

Adding and subtracting $\theta_{0}$ in $\textbf{(1)}$ we get:

$\textbf{(1)} \implies E[\theta'_{MVU}] + (\theta_{0} - \theta_{0}) + E[\alpha \cdot \theta'_{MVU}] - \theta_{0} =$
$0 + \theta_{0} + E[\alpha \cdot \theta'_{MVU}] - \theta_{0} =$
$E[\alpha \cdot \theta'_{MVU}]$

Therefore, for any value of $\alpha \neq 0$, $\theta'_{b}$ 's are **biased** estimators.

### Answer (c)

$MSE(\theta'_{MVU}) = E[(\theta'_{MVU} - E[\theta'_{MVU}])^{2}] + (E[\theta'_{MVU}] - \theta_{0})^{2}$

For MSE to be zero, it would be required that the variance of $\theta'_{MVU}$ is zero. However, in a real setting where the number of samples is finite, exactly because we assume an underlying **sampling distribution** of $\theta'_{MVU}$ which is a **random variable**, we reach the conclusion that this cannot hold since each finite sample will have its own variance.

### Answer (d)

$MSE(\theta'_{b}) = E[(\theta'_{b} - E[\theta'_{b}])^{2}] + (E[\theta'_{b}] - \theta_{0})^{2}$

### Question (b)

Fit a $8^{th}$ degree polynomial on the data using the LS estimator and plot the results (data points and the curve resulting from the fit). Output also the estimates of the parameters of the polynomial.

#### Answer (b)

```python
POLY_ORDER = 8
```

```python
'''
    Returns a nth order polynomially transformed matrix of original X,
    including ones at X[:, 0] so at to be readily available for linear modelling 
    
    :param X: Numpy matrix (X)
    :param order: polynomial order
    
    :returns: Numpy matrix
'''
def poly(X, order = 8):
    # Initialize a X.shape[0]x(order + 1) matrix (nth order poly plus ones)
    X_poly = np.zeros(shape = (X.shape[0], order + 1))
    
    # Ones
    X_poly[:, 0] = 1
    
    # Raise to the i-th power
    for i in range(1, order + 1):
        X_poly[:, i] = np.power(X, i).reshape(X.shape[0])
    
    return X_poly

# Sanity check
assert poly(X, order = POLY_ORDER).shape == (X.shape[0], POLY_ORDER + 1)
```

```python
'''
    Vectorized implementation of computing the estimated Î¸ vector (theta hat).
    
    :param X: Numpy matrix (X)
    :param y: Numpy vector (y)
    
    :returns: Numpy column vector (theta hat)
'''
def compute_theta_hat_ols(X, y):
    # Calculate X.T * X
    XTX = np.dot(X.T, X)

    # Calculate the inverse of X.T * X
    XTX_inv = np.linalg.inv(XTX)
    
    # Calculate theta estimate
    theta_hat_ols = XTX_inv.dot(X.T).dot(y)
    
    return theta_hat_ols

X_dash_ols = poly(X, order = 8)

print("X_dash_ols shape")
print("--------------------------------------------------------------------------")
print(X_dash_ols.shape)

print("\n")

# Get the estimated parameter vector theta
theta_hat_ols = compute_theta_hat_ols(X_dash_ols, y)

print("theta hat OLS shape")
print("--------------------------------------------------------------------------")
print(theta_hat_ols.shape)

theta_hat_ols_names = ["$\theta_0$", "$\theta_1$", "$\theta_2$", "$\theta_3$", "$\theta_4$", "$\theta_5$", "$\theta_6$", "$\theta_7$", "$\theta_8$"]
pd.DataFrame(theta_hat_ols, theta_hat_ols_names, ["Theta OLS coeffiecients"])
```

```python
# Plot the polynomial
fig, ax = plt.subplots(figsize = (12, 6))

# Plot our initial (trianing) data
ax.scatter(X, y, marker = "o", alpha = 0.9, color = "crimson");

# Plot the fitted poly
# Generate 100 values between -0.1 and 1.0
X_sample_ols = np.column_stack([
    np.ones((100, 1)),
    np.linspace(-0.1, 1.1, 100),
    np.linspace(-0.1, 1.1, 100),
    np.linspace(-0.1, 1.1, 100),
    np.linspace(-0.1, 1.1, 100),
    np.linspace(-0.1, 1.1, 100),
    np.linspace(-0.1, 1.1, 100),
    np.linspace(-0.1, 1.1, 100),
    np.linspace(-0.1, 1.1, 100)
])

# Raise to the i-th power
for i in range(1, 9):
    X_sample_ols[:, i] = np.power(X_sample_ols[:, i], i)

y_sample_ols = X_sample_ols.dot(theta_hat_ols)

ax.plot(X_sample_ols[:, 1:2], y_sample_ols, alpha = 0.9, color = "lightslategrey");

# Set plot title
ax.set_title(r"Fitted $8^{th}$ degree polynomial (OLS)", fontdict = plot_main_title_font, pad = 20)

# Set plot axis labels
ax.set_xlabel(r"$X$", fontdict = plot_axis_title_font, labelpad = 20)
ax.set_ylabel(r"$y$", fontdict = plot_axis_title_font, labelpad = 20)

pass;
```

### Question (c)

Fit a $8^{th}$ degree polynomial on the data using the <font color="red">ridge regression</font> estimator and plot the results (data points and the curve resulting from the fit). Output also the estimates of the parameters of the polynomial. Experiment with various values of $\lambda$.

#### Answer (c)

```python
'''
    Vectorized implementation of computing the estimated Î¸ vector (theta hat) 
    for the ridge regression estimator.
    
    :param X: Numpy matrix X
    :param y: Numpy column vector y
    :param l: lambda regularization parameter
    
    :returns: Numpy column vector (theta hat)
'''
def compute_theta_hat_ridge_regression(X, y, l = 0.01):
    # Calculate X.T * X + Î»I
    XTX = np.dot(X.T, X) + (l * np.identity(X.shape[0] - 1))

    # Calculate the inverse of X.T * X
    XTX_inv = np.linalg.inv(XTX)
    
    # Calculate theta estimate
    theta_hat_ridge_regression = XTX_inv.dot(X.T).dot(y)
    
    return theta_hat_ridge_regression

X_dash_ridge_regression = poly(X, order = 8)

print("X_dash_ridge_regression shape")
print("--------------------------------------------------------------------------")
print(X_dash_ridge_regression.shape)

print("\n")

# Get the estimated parameter vector theta
theta_hat_ridge_regression = compute_theta_hat_ridge_regression(X_dash_ridge_regression, y, l = 0.01)

print("theta_hat ridge regression shape")
print("--------------------------------------------------------------------------")
print(theta_hat_ridge_regression.shape)

theta_hat_ridge_regression_names = ["$\theta_0$", "$\theta_1$", "$\theta_2$", "$\theta_3$", "$\theta_4$", "$\theta_5$", "$\theta_6$", "$\theta_7$", "$\theta_9$"]
pd.DataFrame(theta_hat_ridge_regression, theta_hat_ridge_regression_names, ["Theta ridge regression coeffiecients"])
```

Produce <font color="red">15 **noisy** versions</font> of the image adding <font color="red">Gaussian noise</font> with <font color="red">zero mean</font> and <font color="red">variance 256</font>. **In reality, these versions may be different images of exactly the same part of the sky, taken at different times** (to produce a noisy version of <font color="red">A</font>: **(i)** create an <font color="blue">MxN</font> array, <font color="red">B</font>, each one of its entries steming from a zero mean, unit variance normal distribution, **(ii)** multiply the array with â256., **(iii)** the noisy version <font color="red">C</font> of <font color="red">A</font> is produced as <font color="red">C=A+B</font>).

#### Answer (b)

```python
# Compute Gaussian noise
def gaussian_noise(mean = 0, variance = 1, size = (1, 1)):
    return np.random.normal(loc = mean, scale = math.sqrt(variance), size = size)
```

```python
# Create 15 instances of noisy images

# Number of noisy images
K = 15

# MxNxK array of noisy images
# where:
# M: image width in pixels
# N: image height in pixels
# K: number of noisy images
noisy_images = np.zeros((A.shape[0], A.shape[1], K))

# Number of plot columns
plot_cols = 3

# Axis counters
ax_1 = 0
ax_2 = 0

# Setup the plot
fig, ax = plt.subplots(nrows = 5, ncols = plot_cols, figsize = (15, 25))

for i in range(0, K):
    # Compute B
    B = gaussian_noise(mean = 0, variance = 256, size = A.shape)
    
    # Compute C
    C = A + B
    
    # Add C at MxNxi position of the MxNxK array of noisy images
    noisy_images[:, :, i] = C
    
    if(i > 1 and i % plot_cols == 0):
        ax_1 += 1
    ax_2 = (i % plot_cols)
        
    ax[ax_1, ax_2].axis("off")
    ax[ax_1, ax_2].imshow(C, cmap = plt.cm.gray)
    
    # Set graph title
    ax[ax_1, ax_2].set_title(label = "C-" + str(i + 1), loc = "center", fontdict = plot_main_title_font, pad = 10)
```

We observe that the introduction of a few samples of Gaussian noise has introduced higher variance in the image resulting from the difference between the original image and the averaged noisy ones.

Also, visual inspection shows that the averaged noise image appears less noisy than the individual noisy instances created before. Mathematically, let:

$y = f(x_{1}, x_{2})$, the function of our original image data, where $x_{1}, x_{2}$ are the coordinates of a pixel and $y$ its pixel value

$n = g(x_{1}, x_{2})$ the function of Gaussian noise

$z = f(x_{1}, x_{2}) + g(x_{1}, x_{2})$ the operation introducing Gaussian noise in the original image, where $x_{1}, x_{2}$ are the coordinates of a pixel and $z$ its noisy pixel value

$K$ the number of noisy images

Performing the averaging operation accross all $K$ images is represented as:

$\frac{1}{K} \sum_{n = 1}^{K} f(x_{1}, x_{2}) + g(x_{1}, x_{2}) \implies$
$\frac{1}{K} \left( \sum_{n = 1}^{K} f(x_{1}, x_{2}) + \sum_{n = 1}^{K} g(x_{1}, x_{2}) \right) \implies$
$\frac{1}{K} \sum_{n = 1}^{K} f(x_{1}, x_{2}) + \frac{1}{K} \sum_{n = 1}^{K} g(x_{1}, x_{2}) \textbf{ (1)}$

It can be shown that the sample mean $\frac{1}{K} \sum_{n = 1}^{K} g(x_{1}, x_{2})$ is the $\hat{\theta}_{MVU}$ of the Gaussian noise r.v. whose expected value is 0 (by definition). Therefore, $\textbf{(1)}$ becomes $\frac{1}{K} \sum_{n = 1}^{K} f(x_{1}, x_{2}) = f(x_{1}, x_{2})$, which is the  original pixel value.

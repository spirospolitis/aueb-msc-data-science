**Note**: Generic form of the Langrangian

Let:
    
$$f(x, y) \text{: the function that we wish to minimize / maximize}$$

$$g(x, y) = \rho \implies g(x, y) - \rho = 0 \text{: the contraint function}$$

The Langrangian function is then formulated as:

$$L(x, y, \lambda) = f(x, y) + \lambda (g(x, y))$$

where $\lambda$ is the *Langrange multiplier*.

## Exercise 1

### Answer

Formulating our problem as the Langragian function:

$$L(\mu) = \sum_{i=1}^{N} (x_{n} - \mu)^{2} + \lambda((\mu - \mu_{0}) - \rho)$$

then, taking the derivative w.r.t $\mu$:

$$\frac{\partial L(\mu)}{\partial \mu} =$$

$$-2\sum_{n=1}^{N} \left( x_{n} - \mu \right) + 2\lambda(\mu - \mu_{0}) =$$

$$-2\sum_{n=1}^{N} + 2N\mu + 2\lambda\mu - 2\lambda\mu_{0}$$

and equating to zero, we finally get $\hat{\mu}_{RR}$:

$$\frac{\partial L(\mu)}{\partial \mu} = 0 \implies$$

$$-2\sum_{n=1}^{N} x_{n} + 2N\mu + 2\lambda\mu - 2\lambda\mu_{0} = 0 \implies$$

$$2 \left( -\sum_{n=1}^{N} x_{n} + N\mu + \lambda\mu - \lambda\mu_{0} \right) = 0 \implies$$

$$-\sum_{n=1}^{N} x_{n} + N\mu + \lambda\mu - \lambda\mu_{0} = 0 \implies$$

$$\mu(N + \lambda) = \sum_{n=1}^{N} x_{n} + \lambda\mu_{0} \implies$$

$$\hat{\mu}_{RR} = \frac{\sum_{n=1}^{N} x_{n} + \lambda\mu_{0}}{N + \lambda}$$

## Exercise 2

### Answer

We define the Langragian function:

$$L(P_{1}, P_{2}, \dots, P_{m}) = \sum_{i=1}^{N} \sum_{j=1}^{m} P(j|x_{i}) ln(P_{j}) + \lambda \left( \sum_{j=1}^{m} P_{j} - 1 \right) \textbf{ (1)}$$

Expanding all $P_{j}$ terms, $\textbf{(1)}$ can be rewritten as:

$$L(P_{1}, P_{2}, \dots, P_{m}) = \sum_{i=1}^{N} P(1|x_{i}) ln(P_{1}) + P(2|x_{i}) ln(P_{2}) + \dots + P(m|x_{i}) ln(P_{m}) + \lambda \left( P_{1} + P_{2} + \dots + P_{m} - 1 \right) \textbf{ (2)}$$

Taking the derivatives $j = 1, \dots, m$, w.r.t $P_{j}$ (from $\textbf{(2)}$) we get the following forms:

$j = 1$

$$\frac{\partial L(P_{1}, P_{2}, \dots, P_{m})}{\partial P_{1}} =
\sum_{i=1}^{N} \frac{P(1|x_{i})}{P_{1}} + \lambda
$$

$j = 2$

$$\frac{\partial L(P_{1}, P_{2}, \dots, P_{m})}{\partial P_{2}} =
\sum_{i=1}^{N} \frac{P(2|x_{i})}{P_{2}} + \lambda
$$

$$\vdots$$

$j = m$

$$\frac{\partial L(P_{1}, P_{2}, \dots, P_{m})}{\partial P_{m}} =
\sum_{i=1}^{N} \frac{P(m|x_{i})}{P_{m}} + \lambda
$$

The functions above provide us with the following form of the partial derivative of the Langrangian w.r.t. $P_{m}$:

$$\sum_{i=1}^{N} \frac{P(j|x_{i})}{P_{j}} + \lambda \textbf{ (3)}$$

Equating $\textbf{(3)}$ to zero we get:

$$\sum_{i=1}^{N} \frac{P(j|x_{i})}{P_{j}} + \lambda = 0 \implies$$

$$\frac{1}{P_{j}} + \sum_{i=1}^{N} P(j|x_{i}) = -\lambda \implies$$

$$P_{j} = -\frac{\sum_{i=1}^{N} P(j|x_{i})}{\lambda} \textbf{ (4)}$$

We have thereby expressed $P_{j}$ w.r.t. $\lambda$.

The constraint equation is:

$$\sum_{j=1}^{m} P_{j} = 1 \textbf{ (5)}$$

Substituting $\textbf{(4)}$ in $\textbf{(5)}$ and expanding all $j$ terms we get:

$$\sum_{j=1}^{m} -\frac{\sum_{i=1}^{N} P(j|x_{i})}{\lambda} = 1 \implies$$
$$-\frac{1}{\lambda} \left( \sum_{i=1}^{N} P(1|x_{i}) + \sum_{i=1}^{N} P(2|x_{i}) + \dots + \sum_{i=1}^{N} P(m|x_{i}) \right) = 1 \implies$$
$$\lambda = - \sum_{i=1}^{N} \left[ P(1|x_{i}) + P(2|x_{i}) + \dots + P(m|x_{i}) \right] \implies$$
$$\lambda = - \sum_{i=1}^{N} \sum_{j=1}^{m} P(j|x_{i}) \textbf{ (6)}$$

From $\textbf{(3)}$ and $\textbf{(5)}$ we get:

$$P_{j} = -\frac{\sum_{i=1}^{N} P(j|x_{i})}{-\sum_{i=1}^{N} \sum_{j=1}^{m} P(j|x_{1})} \implies$$

$$P_{j} = \frac{\sum_{i=1}^{N} P(j|x_{i})}{\sum_{i=1}^{N} \sum_{j=1}^{m} P(j|x_{1})} \textbf{ (7)}$$

Because we know that $\sum_{j=1}^{m} P(j|x_{1}) = 1$ (slide 4 in "Machine Learning and Computational Statistics", Unit 6 notes), $\textbf{(7)}$ is expressed as:

$$P_{j} = \frac{\sum_{i=1}^{N} P(j|x_{i})}{\sum_{i=1}^{N} 1} \implies$$

$$P_{j} = \frac{1}{N} \sum_{i=1}^{N} P(j|x_{i})$$

Intuitively, we notice that observations from data set $_{1}$ come from a multivariate Gaussian distribution, no visible clusters can be identified.

We should proceed with estimating the PDF from which the observations come from.

The mean vector of $X_{1}$ is calculated as:

```python
X_1_mu = np.mean(X_1, axis = 0).reshape(1, X_1.shape[1])
```

Centering the data points to their mean:

```python
X_1_centered = X_1 - X_1_mu
```

We can visualize the distribution of data and estimate the PDF with the use of Seaborn:

```python
# Plot
fig, ax = plt.subplots(figsize = (18, 8))

# Plot observations
# sns.distplot(X_1_centered[:, 0], hist = True, kde = True, bins = int(len(np.unique(X_1[:, 0])) / 10), color = "crimson", hist_kws = {"color": "lightblue", "edgecolor": "grey"}, kde_kws = {"linewidth": 4}, label="X1")
# sns.distplot(X_1_centered[:, 1], hist = True, kde = True, bins = int(len(np.unique(X_1[:, 1])) / 10), color = "crimson", hist_kws = {"color": "lightblue", "edgecolor": "grey"}, kde_kws = {"linewidth": 4}, label="X2")
# sns.distplot(X_1_centered[:, 2], hist = True, kde = True, bins = int(len(np.unique(X_1[:, 2])) / 10), color = "crimson", hist_kws = {"color": "lightblue", "edgecolor": "grey"}, kde_kws = {"linewidth": 4}, label="X2")
# sns.distplot(X_1_centered[:, 3], hist = True, kde = True, bins = int(len(np.unique(X_1[:, 3])) / 10), color = "crimson", hist_kws = {"color": "lightblue", "edgecolor": "grey"}, kde_kws = {"linewidth": 4}, label="X2")

# Equivalent to the above commented-out code
sns.distplot(X_1_centered[:, 0] + X_1_centered[:, 1] + X_1_centered[:, 2] + X_1_centered[:, 3], hist = True, kde = True, bins = 100, color = "crimson", hist_kws = {"color": "lightblue", "edgecolor": "grey"}, kde_kws = {"linewidth": 4})

# Set plot title
ax.set_title(r"Histogram and PDF estimation for $X_{1}$", fontdict = plot_main_title_font, pad = 20)

# Set plot legend
label_lightblue = patches.Patch(color = "lightblue", label = "Histogram")
label_crimson = patches.Patch(color = "crimson", label = "PDF estimation")
ax.legend(handles = [label_lightblue, label_crimson])

pass;
```

We notice that the observations seem to be generated by a multivariate Gaussian, i.e. $N \sim (\boldsymbol{\mu}, \Sigma)$. Note, however, that we should perform normality tests to identify whether our data is trully normal.

The parameters of the multivariate Gaussian are the mean vector $\boldsymbol{\mu}$ (already calculated) and the covariance matrix $\Sigma$, calculated below:

```python
X_1_sigma = np.matmul(X_1_centered.T, X_1_centered) / len(X_1_centered)
```

Having estimated the parameters of the multivariate normal, we can proceed with evaluating the probabilities of the test data set:

```python
X_1_test = np.array([
    [2.01, 2.99, 3.98, 5.02],
    [20.78, -15.26, 19.38, -25.02],
    [3.08, 3.88, 4.15, 6.02]
])
```

```python
X_1_test_prob = multivariate_normal.pdf(X_1_test, mean = X_1_mu[0], cov = X_1_sigma)
X_1_names = np.array(["$X$", "$P(X)$"])
X_1_df = pd.DataFrame([
    [X_1_test[0], "{0:1.4f}".format(X_1_test_prob[0])],
    [X_1_test[1], "{0:1.4f}".format(X_1_test_prob[1])],
    [X_1_test[2], "{0:1.4f}".format(X_1_test_prob[2])]
], columns = X_1_names)
X_1_df
```

The above table indicates that all test data points are at the edges of the multivariate Gaussian, all with very low probability of occurence.

Visual inspection of the training observations along with the test data points is provided below:

```python
from mpl_toolkits.mplot3d import Axes3D

fig, ax = plt.subplots(figsize = (14, 14), sharey = True)
ax = plt.axes(projection = "3d")
ax.view_init(elev = 25, azim = 45)

# Plot X_1 data points
ax.scatter(X_1[:, 0], X_1[:, 1], X_1[:, 2], c = X_1[:, 3], marker = "o", alpha = 0.05)

# Plot X_1 test data points
ax.scatter(X_1_test[:, 0], X_1_test[:, 1], X_1_test[:, 2], s = 200, c = "crimson", marker = "+", alpha = 1)

# Set plot title
ax.set_title("Visualization of $X_{1}$ with test data points", fontdict = plot_main_title_font, pad = 40)

# Set plot axis labels
ax.set_xlabel(r"$x_{1}$", fontdict = plot_axis_title_font, labelpad = 10)
ax.set_ylabel(r"$x_{2}$", fontdict = plot_axis_title_font, labelpad = 10)
ax.set_zlabel(r"$x_{3}$", fontdict = plot_axis_title_font, labelpad = 10)

pass;
```
